# Search

```python
import numpy as np
```

```python
var = np.array([1,5,4,2,3,6,7,9,8])
x = np.where(var == 2)
print(x)# return array with element is index of result
```

```python
var = np.array([1,5,4,2,3,6,7,9,8])
x = np.where((var%2) == 0)
print(x)# return array with element is devide 2 equal 0
```

```python
var1 = np.array([1,2,3,4,8,9])
x1 = np.searchsorted(var1, [5,6,7],side="right")
print(x1)
```

# Sort

```python
var1 = np.array([2,5,1,3,4,8,6,9,7])
print(np.sort(var1))
```

```python
var2 = np.array(["a","s","d","f"])
print(np.sort(var2))
```

```python
var2 = np.array([[4,2,3],[1,12,5],[22,52,6]])
print(np.sort(var2))
```

# Filter

```python
var = np.array(["a","s","d","f"])
f = [True, False, False, True]
new_a = var[f]
print(new_a)
print(type(new_a))
```

# Shuffle

```python
var = np.array([1,2,3,4,5])
np.random.shuffle(var)
print(var)
```

# Unique

```python
var1 = np.array([1,2,3,4,2,5,2,6,2,7])
x = np.unique(var1, return_index=True,return_counts=True)
print(x)
```

# Resize

```python
var = np.array([1,2,3,4,5,6])
y = np.resize(var,(2,3))
y
```

```python
y = np.resize(var,(3,2))
y
```

# Flatten

```python
var = np.array([[1,2],[3,4],[5,6]])
print(var.flatten(order="C"))
print(var.flatten(order="F"))
print(var.flatten(order="A"))
print(var.flatten(order="K"))
```

# Ravel

```python
var = np.array([[1,2],[3,4],[5,6]])
print(np.ravel(var,order="F"))
```
