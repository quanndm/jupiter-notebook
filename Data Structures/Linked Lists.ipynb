```python
class Node:
    def __init__(self,data=None, next=None):
        self.data = data
        self.next = next
```

```python
class LinkedList:
    def __init__(self):
        self.head = None
    
    def insert_at_begining(self,data):
        self.head = Node(data, self.head)
        
    def insert_at_end(self,data):
        if self.head == None:
            self.head = Node(data, self.head)
            return
        itr = self.head
        while itr.next:
            itr = itr.next
        
        itr.next = Node(data, None)
    def print(self):
        if self.head == None:
            print("Linked list is empty!")
            return
        itr = self.head
        result_str = ""
        while itr:
            if itr.next != None:
                result_str += str(itr.data) + " --> "
            else:
                result_str += str(itr.data)
            itr = itr.next
            
        print(result_str)
    
    def insert_values(self, data_list):
        self.head = None
        for data in data_list:
            self.insert_at_end(data)
    
    def get_length(self):
        count = 0
        itr = self.head
        while itr:
            count += 1
            itr = itr.next
        return count
    def remove_at(self, index):
        if index < 0 or index >= self.get_length():
            raise Exception("Invalid Index!")
        if index == 0:
            self.head = self.head.next
            return
        count = 0
        itr = self.head
        while itr:
            if count == index-1:
                itr.next = itr.next.next
                break
            count += 1 
            itr = itr.next      
    def insert_at(self, data, index):
        if index < 0 or index > self.get_length():
            raise Exception("Invalid Index!")
        if index == 0:
            self.insert_at_begining(data)
            return
        count = 0
        itr = self.head
        while itr:
            while count == index -1:
                node = Node(data, itr.next)
                itr.next = node
                break
            count += 1
            itr = itr.next
            
    def insert_after_value(self, data_after, data_insert):
        # find index item first occur
        index = self.search(data_after)
        if index == -1 :
            return False
        # insert item after index
        if index == self.get_length():
            self.insert_at_end(data_insert)
            return True
        self.insert_at(data_insert, index+1)
        return True
        
        
    def remove_by_value(self, data_remove):
        #remove first item occur
        index = self.search(data_remove)
        if index == -1 :
            return False
        self.remove_at(index)
        return True
    def search(self, data):
        index = -1
        itr = self.head
        count = 0
        while itr:
            if data == itr.data:
                index = count
                break
            count += 1
            itr = itr.next
        return index
```
