```python
class BinarySearchTreeNode:
    def __init__(self,data):
        self.data = data
        self.left = None
        self.right = None
    def add_child(self, data):
        if data == self.data:
            return
        if data < self.data:
            # add data in left subtree
            if self.left:
                self.left.add_child(data)
            else:
                self.left = BinarySearchTreeNode(data)
        else:
            # add data in right subtree
            if self.right:
                self.right.add_child(data)
            else:
                self.right = BinarySearchTreeNode(data)
    def pre_order_traversal(self):
        # duyet cay nhi phan
        elements = []
        # Visit base Node
        elements.append(self.data)
        # Visit left tree
        if self.left:
            elements += self.left.in_order_traversal()
        # Visit right tree
        if self.right:
            elements += self.right.in_order_traversal()
        return elements
    def post_order_traversal(self):
        # duyet cay nhi phan
        elements = []
        # Visit left tree
        if self.left:
            elements += self.left.in_order_traversal()
        # Visit right tree
        if self.right:
            elements += self.right.in_order_traversal()
        # Visit base Node
        elements.append(self.data)
        
        return elements
    def in_order_traversal(self):
        # duyet cay nhi phan
        elements = []
        # Visit left tree
        if self.left:
            elements += self.left.in_order_traversal()
        # Visit base Node
        elements.append(self.data)
        # Visit right tree
        if self.right:
            elements += self.right.in_order_traversal()
        return elements
    def search(self, val):
        if self.data == val:
            return True
        if val < self.data:
            # value may be in left subtree
            if self.left:
                return self.left.search(val)
            else:
                return False
        if val > self.data:
            # value may be in right subtree
            if self.right:
                return self.right.search(val)
            else:
                return False
    def find_min(self):
        # find min element in tree
        if self.left is None:
            return self.data
        return self.find_min()
    
    def find_max(self):
        #find max element in tree
        if self.right is None:
            return self.data
        return self.find_max()
    
    def calculate_sum(self):
        left_sum = self.calculate_sum() if self.left else 0
        right_sum = self.calculate_sum() if self.right else 0
        return self.data + left_sum + right_sum
    #check  ######
    def height(self):
        if not self.left or not self.right:
            return 0
        else:
            # Compute the height of each subtree
            l_height = self.left.height()
            r_height = self.right.height()
            
            # Use the larger one
            if l_height > r_height:
                return l_height+1
            else:
                return r_height+1
    def print_current_level(self, level):
        if not self.left or not self.right:
            return
        if level == 1:
            print(self.data, end=" ")
        elif level > 1:
            self.left.print_current_level(level-1)
            self.right.print_current_level(level-1)
    def print_level_order(self):
        h = self.height()
        for i in range(1, h+1):
            self.print_current_level( i)
    #### check
    def delete(self,val):
        # 1. Delete Node with no child
        # 2. Delete Node one child
        # 3. Delete Node with two children
        # ....
        if val < self.data:
            if self.left:
                self.left = self.left.delete(val)
        elif val > self.data:
            if self.right:
                self.right = self.right.delete(val)
        else:
            if self.left is None and self.right is None:
                return None
            if self.left is None:
                return self.right
            if self.right is None:
                return self.right
            
            min_val = self.right.find_min()
            self.data = min_val
            
            self.right = self.right.delete(min_val)
        return self
```

```python
def build_tree(elements):
    root = BinarySearchTreeNode(elements[0])
    for i in range(1, len(elements)):
        root.add_child(elements[i])
        
    return root
```
